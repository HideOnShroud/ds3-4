```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Earthquakes_database.csv")
df.head()
```

```python
df[['Month', 'Day', 'Year']] = df['Date'].str.split('/', expand=True)
df.head()
```

```python
df.drop(columns="Date", inplace=True)
```

```python
df.head()
```

```python
bins = [0, 5.5, 6.0, 7.0] 
labels = ['Low', 'Moderate', 'High']  

df['Magnitude Category'] = pd.cut(df['Magnitude'], bins=bins, labels=labels, right=False)
df.head()
```

```python
pivot_table_df = pd.pivot_table(
    df,
    values='Magnitude', 
    index='Year',       
    columns='Status',    
    aggfunc='mean'      
)
pivot_table_df
```

```python
df_stack = df.stack()
df_stack
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 8))
magnitude_counts = df['Magnitude Category'].value_counts()
plt.pie(magnitude_counts, labels=magnitude_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Earthquakes by Magnitude Category')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Depth', 
    y='Magnitude', 
    hue='Magnitude Category', 
    palette='viridis', 
    data=df, 
    s=100
)
plt.title('Scatter Plot of Depth vs. Magnitude')
plt.xlabel('Depth (km)')
plt.ylabel('Magnitude')
plt.legend(title='Magnitude Category')
plt.show()
```

```python
df = df.sort_values('Time')

plt.figure(figsize=(12, 6))
sns.lineplot(x='Time', y='Magnitude', data=df, marker='o', linestyle='-', color='b')
plt.title('Earthquake Magnitudes Over Time')
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Magnitude Category', y='Depth', data=df, estimator=sum)
plt.title('Total Depth of Earthquakes by Magnitude Category')
plt.xlabel('Magnitude Category')
plt.ylabel('Total Depth')
plt.show()
```

```python
df['Time'] = pd.to_datetime(df['Time'])

monthly_data = df.resample('M').agg({'Magnitude': 'mean', 'Depth': 'mean'})
monthly_data
```
